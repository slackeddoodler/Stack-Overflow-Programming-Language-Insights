```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data/survey_results_public.csv', index_col='Respondent')
pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 100)
```

```python
df.head(25)
```

```python
filtered_data = df['LanguageWorkedWith'].dropna()

split_lists = filtered_data.apply(lambda x: x.split(';'))
```

```python
from itertools import chain
all_languages = list(chain.from_iterable(split_lists))
lang_series = pd.Series(all_languages)
```

```python
languages = lang_series.value_counts().index.tolist()
values = lang_series.value_counts().values.tolist()

plt.figure(figsize=(12, 10))

bars = plt.barh(languages, values)

plt.title('What are the Top Languages?')

plt.bar_label(bars, padding = 3, labels=[f"{int(v/1000)}K" for v in values])

plt.gca().invert_yaxis()

plt.tight_layout()

plt.savefig('What are the Top Langauges.png')

plt.show()
```
